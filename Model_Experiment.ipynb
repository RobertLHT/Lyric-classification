#### Data preprocessing 

Text data is not numerical value, if we want to use it as feature in the ML or RNN nerual network, we need to convert it to the numerical value. And it generally follows the following procedure: 
1. Text processing (lower case, remove numbers)

2. Tokenize (Change sentence to a word list) 

3. Word embedding (Hash word to numerical value)

There are many useful package used for text data preprocessing such Tokenizer from tensorflow, it does not simply change sentence to the word, it also calcualtes the frequency of words. And WordCloud package we used to display the frequent words.<br>

In the following experiment, we will use countVectorizer for ML model, because we test the TF-IDF does not perform vell as the countVectorizer in our experiment

```python
# Get train and test data for training process
X_train, X_test, y_train, y_test = train_test_split(train.drop(["Genre"],axis = 1), train["Genre"], test_size=0.3, random_state = 5)

from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(ngram_range=(1,2), stop_words="english")
vectorizer.fit(train["Lyric"])
X_train_cv = vectorizer.transform(X_train["Lyric"])
X_test_cv = vectorizer.transform(X_test["Lyric"])
```

#### Hyperparameter tuning

We use Gridsearchcv to find the best parameter of the Naive Bayes model. Because it will automaticlly use cross validation, so we will entire training data set for training process. In addition, we also want to try different parameters of the countVectorizer, so we set up a pipline of our model, and then use grid search to find the best hyperparameters.

```python
X, y= train["Lyric"], train["Genre"]

# Define a Pipline for the model
pipe_model = Pipeline([('cv', CountVectorizer(stop_words="english")),
             ('model', MultinomialNB())
            ])

parameters = {
  "cv__ngram_range":[(1,1),(1,2),(2,2)],
  "model__fit_prior":[True, False],
}

gs_model = GridSearchCV(pipe_model,parameters,cv=5, verbose=1)
gs_model.fit(X, y)
```

```python
print(f"The best accuracy is {round(gs_model.best_score_*100,2)}%")
print("The best parameter is ",gs_model.best_params_)
```

### RNN nerual network

After we used ML model as classifier, we also want to try RNN network for prediction.

#### Data preprocessing

Because the nerual network won't take categorical value as input or output, we need to preprocess the features and label. First we need to encode the label

```python
df = pd.read_csv("/content/drive/MyDrive/2179/project/data.csv")

train = df.head(50000)
le = preprocessing.LabelEncoder()
le.fit(train.Genre)
train["Genre"] = le.transform(train["Genre"])
print("Encode label to a integer")
print("The first five label are \n", train["Genre"][:5].to_string(index = False))

X = train.drop(["Genre"], axis = 1)
y= to_categorical(train["Genre"],3)
print("After one hot coding")
print("The first five label are \n", y[:5])
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.25, random_state = 5)
```

We need to preprocess the text feature for training

```python
#https://www.tensorflow.org/text/tutorials/text_classification_rnn
train_dataset = tf.data.Dataset.from_tensor_slices((X_train, y_train))
test_dataset = tf.data.Dataset.from_tensor_slices((X_valid, y_valid))



BATCH_SIZE = 64
train_dataset = train_dataset.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
test_dataset = test_dataset.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

VOCAB_SIZE = 1000
encoder = tf.keras.layers.TextVectorization(max_tokens= VOCAB_SIZE,
     ngrams = (1,2))
encoder.adapt(train_dataset.map(lambda text, label: text))
# vocab = np.array(encoder.get_vocabulary())
```

We only take 1000 vocabulary for the training process, and still we use n_gram (1,2) because we have found it has better performance on previous experiment

#### Define the model

We use GRU for the RNN, because we tested both lstm and gru. Gru is fast and has better performance. In addition, we implement early stop to aviod overfitting

```python
early_stop = EarlyStopping(monitor='val_loss',patience=3)
```

```python
gru = Sequential()
gru.add(encoder)
gru.add(Embedding(input_dim = VOCAB_SIZE, output_dim=64, mask_zero=True))
gru.add(GRU(256))
# gru.add(Flatten())

gru.add(Dense(64,activation = "relu"))
gru.add(Dropout(0.2))
gru.add(Dense(32,activation = "relu"))
gru.add(Dropout(0.2))
gru.add(Dense(3, activation = "softmax"))
gru.compile(loss='categorical_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])

history = gru.fit(train_dataset,epochs=15,
          validation_data=test_dataset,callbacks=[early_stop])
```

We see that the performance is still around 66%, we can plot these loss and accuracy.

```python
plt.title("Training History")
plt.xlabel("Epoch")
plt.ylabel("Performance")
plt.plot(history.history["val_accuracy"], label = "val_accuracy")
plt.plot(history.history["accuracy"], label = "accuuracy")
plt.plot(history.history["val_loss"], label = "val_loss")
plt.plot(history.history["loss"], label = "loss")
plt.legend()
plt.show()
```

We see that around 7th epoch, the model start overfitting, which the val_loss surpass the loss, and val accuracy stop increase. However, we still believe the model is still valid as acceptable classifier. Let's check the confusion matrix of the valid data and precsion of the valid dataset

```python
pred_valid = gru.predict(X_valid)
pred_valid = np.argmax(pred_valid,axis = 1)
pred_valid_label = le.inverse_transform(pred_valid)

y_valid = np.argmax(y_valid,axis = 1)
y_valid = le.inverse_transform(y_valid )
```

```python
cm = confusion_matrix(y_valid, pred_valid_label, labels = ["Rock", "Pop", "Hip Hop"])
cm_df = pd.DataFrame(cm, index = ["Rock", "Pop", "Hip Hop"], columns = ["Rock", "Pop", "Hip Hop"])

plt.figure(figsize=(5,4))
sns.heatmap(cm_df, annot=True,fmt='d')
plt.title('Confusion Matrix')
plt.ylabel('Actal Values')
plt.xlabel('Predicted Values')
plt.show()
```

```python
detail_result = pd.DataFrame(precision_recall_fscore_support(y_valid, pred_valid_label, labels =["Rock", "Hip Hop", "Pop"])[:3], index = ["Precision", "Recall", "F1-score"], columns = ["Rock", "Pop", "Hip Hop"])
plt.figure(figsize=(5,4))
sns.heatmap(detail_result, annot=True)
plt.title('Detailed result')
plt.ylabel('Score')
plt.xlabel('label')
plt.show()
```
